```python
val_df = pd.read_csv('malayalam_hope_dev.csv', header=None, delimiter='\t')
```

```python
val_df = val_df.drop([2], axis = 1) 
val_df = val_df.rename(columns={0: "text", 1: "label"})
val_df.head()
```

```python
!pip install langdetect
!pip install demoji
import demoji
demoji.download_codes()
!pip install google_trans_new
from langdetect import detect
from google_trans_new import google_translator
```

REMOVING EMOJIS


```python
translator = google_translator()
for index, row in val_df.iterrows():
  # Check if emoji is contained in text
  text = str(row['text'])
  found = demoji.findall(text)
  if found:
    # There is an emoji now detect language
    if detect(text) == 'ml':
      print(text)
      # Text is Malayalam
      for key, value in found.items():
        emojiEng = value
        # Translate to Malayalam
        transText = translator.translate(emojiEng, lang_tgt='ml', lang_src='en')
        text = text.replace(key, transText + ' ')
    else:
      # Text is english
      for key, value in found.items():
        text = text.replace(key, value + ' ')
    
    row['text'] = text
    print(text)
    print("\n")
```

```python
for index, row in val_df.iterrows():
  text = str(row['text'])
  found = demoji.findall(text)
  if(found):
    print('EMOJI FOUND')
```

REMOVE CONTRACTIONS

```python
# import library
!pip install contractions
import contractions
# contracted text
for index, row in val_df.iterrows():
  # creating an empty list
  expanded_words = []	
  for word in str(row['text']).split():
  # using contractions.fix to expand the shotened words
    expanded_words.append(contractions.fix(word))
    
  expanded_text = ' '.join(expanded_words)
  row['text'] = expanded_text
```

REMOVE PUNCTUATIONS

```python
import string
for index, row in val_df.iterrows():   
  # An input string.
  Sentence = str(row['text'])
  for i in Sentence:
      # checking whether the char is punctuation.
      if i in string.punctuation:
        Sentence = Sentence.replace(i, '')
  row['text'] = Sentence.lower()
```

REMOVING STOPWORDS

```python
new_df = pd.DataFrame(columns=['text','label'])
new_df
```

```python
!pip install advertools
import advertools as adv
```

```python
eng_stop_words=adv.stopwords['english']

for index, row in val_df.iterrows():
  sent = ''
  end = {}
  line=str(row['text'])
  for i in line.split():
    if i not in eng_stop_words :
      sent+=i + " "
  end['text']=sent
  end['label']=str(row['label'])
  new_df=new_df.append(end,ignore_index=True)
  #     end.append(i)
  # row['text'] = (" ").join(end)

new_df.head()
```
